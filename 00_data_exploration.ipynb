# Data Exploration: Drill CSV Files

## Setup and Dependencies

```python
# Check if required packages are installed
import importlib.util

packages_to_check = ['pandas']
missing_packages = []

for package in packages_to_check:
    if importlib.util.find_spec(package) is None:
        missing_packages.append(package)

if missing_packages:
    print(f"⚠️  Missing packages: {', '.join(missing_packages)}")
    print(f"Install with: pip install {' '.join(missing_packages)}")
else:
    print("✓ All required packages are installed")
```

```python
# Import required libraries
import pandas as pd
from pathlib import Path

# Define paths
DATA_PATH = Path.home() / "Downloads"

# Define CSV file paths
csv_files = {
    'warmup': DATA_PATH / "warmup_drills.csv",
    'cool_down': DATA_PATH / "cool_down_drills.csv",
    'conditioning_with_ball': DATA_PATH / "conditioning_with_ball.csv",
    'small_sided_games': DATA_PATH / "small_sided_games.csv",
    'tactical': DATA_PATH / "tactical_drills.csv",
    'technical': DATA_PATH / "technical_drills.csv"
}

# Verify all files exist
missing_files = [name for name, path in csv_files.items() if not path.exists()]

if missing_files:
    raise FileNotFoundError(
        f"❌ Missing CSV file(s): {', '.join(missing_files)}\n"
        f"Please ensure all drill CSV files are in your Downloads folder."
    )

print(f"✓ Found all {len(csv_files)} CSV files in {DATA_PATH}")
```

## Preview Each CSV File

```python
# Load and preview each CSV file
for category, file_path in csv_files.items():
    print(f"\n{'='*70}")
    print(f"📁 {category.upper().replace('_', ' ')}")
    print(f"{'='*70}")
    
    # Read the CSV
    df = pd.read_csv(file_path)
    
    # Display basic info
    print(f"\nShape: {df.shape[0]} rows × {df.shape[1]} columns")
    print(f"\nColumns: {df.columns.tolist()}")
    print(f"\nFirst 5 rows:")
    display(df.head())
    print("\n")
```

## Combine All CSV Files

```python
# Combine all CSV files into a single DataFrame with category labels
all_drills = []

for category, file_path in csv_files.items():
    df = pd.read_csv(file_path)
    df['category'] = category
    all_drills.append(df)

# Concatenate all dataframes
drills_raw = pd.concat(all_drills, ignore_index=True)

# Assign duration to each drill based on category
duration_map = {
    'warmup': 15,
    'cool_down': 15,
    'conditioning_with_ball': 15,
    'small_sided_games': 25,
    'tactical': 20,
    'technical': 25
}

drills_raw['duration'] = drills_raw['category'].map(duration_map)

print(f"✓ Combined dataset: {len(drills_raw)} total drills from {len(csv_files)} categories")
print(f"\nDrills by category:")
print(drills_raw['category'].value_counts())
print(f"\nCombined shape: {drills_raw.shape[0]} rows × {drills_raw.shape[1]} columns")
print(f"\nDuration by category:")
print(drills_raw.groupby('category')['duration'].first())

# Preview combined dataset
drills_raw
```

## Data Quality Checks

```python
# Inspect data types and non-null counts
print("Dataset Info:")
drills_raw.info()
```

```python
# Calculate missingness by column (sorted descending)
missingness = drills_raw.isna().mean().sort_values(ascending=False)

print("Missing Value Percentages (sorted by highest missingness):\n")
print(missingness[missingness > 0])  # Only show columns with missing values

if missingness[missingness > 0].empty:
    print("✓ No missing values detected in any column")
```

```python
# Check for list-like strings that will need parsing
# These often appear as: "['value1', 'value2', 'value3']"

print("Checking for list-like string patterns...\n")

for col in drills_raw.columns:
    # Check if column contains string data
    if drills_raw[col].dtype == 'object':
        # Sample first non-null value
        sample = drills_raw[col].dropna().iloc[0] if len(drills_raw[col].dropna()) > 0 else None
        
        if sample and isinstance(sample, str):
            # Check if it looks like a list string
            if sample.strip().startswith('[') and sample.strip().endswith(']'):
                print(f"⚠️  '{col}' contains list-like strings")
                print(f"    Example: {sample}")
                print()
```

```python
# Check numeric columns for min/max players, duration, etc.
numeric_cols = drills_raw.select_dtypes(include=['int64', 'float64']).columns

if len(numeric_cols) > 0:
    print("Numeric column summary:\n")
    print(drills_raw[numeric_cols].describe())
else:
    print("⚠️  No numeric columns detected - may need type conversion")
```
